```python
from collections import defaultdict
```

```python
def group_anagrams(words):
    anagram_groups = defaultdict(set)
    
    for word in words:
        key = ''.join(sorted(word))
        anagram_groups[key].add(word)
    
    return list(anagram_groups.values())
```

```python
# Take user input for the list of words
word_list = input("Enter a list of words, separated by spaces: ").split()

# Call the group_anagrams function with the user input
result = group_anagrams(word_list)

# Print the result
print(result)
```

```python
#testing
import random
import string

N = 10000  # Number of words
M = 100  # Maximum length of each word
words = [''.join(random.choice(string.ascii_lowercase) for _ in range(random.randint(1, M))) for _ in range(N)]

result = group_anagrams(words)
print(result)
```
